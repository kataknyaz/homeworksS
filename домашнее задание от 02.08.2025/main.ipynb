Часть 1. Виды сайтов и технологии рендеринга
Типы сайтов с примерами:
Статические (например, лендинги, сайты-визитки) — контент неизменен без ручного редактирования кода.
Динамические (например, интернет-магазины, соцсети) — контент генерируется на основе данных из БД.
SPA (Single Page Application)(например, Gmail, Trello) — загружаются один раз, обновляются без перезагрузки страницы.
SSR (Server-Side Rendering):
Принцип работы: HTML генерируется на сервере и отправляется клиенту в готовом виде.
Преимущества: Лучше для SEO, быстрая начальная загрузка, работает на слабых устройствах.
Гибридный подход (SSR + SPA):
Суть: Например, Next.js или Nuxt.js. Первая загрузка через SSR, далее SPA-логика.
Преимущества: SEO-дружественность + интерактивность SPA.
MPA (Multi-Page Application):
Отличие от SPA: Каждая страница загружается с сервера отдельно, что проще для SEO, но менее интерактивно.

Часть 2. SEO и влияние ИИ
SEO (Search Engine Optimization):
Определение: Оптимизация сайта для повышения видимости в поисковых системах.
Важность: Привлекает органический трафик, увеличивает конверсию.
Виды SEO:
Техническое: Оптимизация кода, скорости загрузки.
Контентное: Качество и релевантность контента.
Офф-страничное: Внешние факторы (ссылки, соцсети).
Влияние ИИ:
Ранжирование: Алгоритмы (например, BERT) лучше понимают контекст.
Генерация контента: ИИ создаёт тексты, но требует проверки.
Анализ данных: Автоматизация аудита и прогнозирования трендов.

Часть 3. Системный дизайн
System Design:
Определение: Проектирование архитектуры сложных систем (например, YouTube, Twitter).
Роль: Позволяет учесть масштабируемость, отказоустойчивость, производительность.
Этапы:
Сбор требований.
Выбор компонентов (API, БД, кеширование).
Оптимизация (например, шардирование БД).

Часть 4. Методологии разработки
Методология разработки:
Определение: Набор практик для управления процессом создания ПО.
Agile:
Принципы: Гибкость, итеративность, вовлечение клиента.
Преимущества: Быстрая адаптация к изменениям.
RAD vs Waterfall:
RAD: Быстрые прототипы, клиент участвует на каждом этапе.
Waterfall: Жёсткая последовательность этапов, изменения сложны.
