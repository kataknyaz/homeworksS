- Затем напишите программу, которая будет запрашивать ввод чисел у

8.2. Практическая часть
- Создайте лямбда-функцию, которая принимает число и возвращает его
куб
- Примените эту лямбда-функцию к каждому элементу списка [1, 2, 3, 4, 5],
используя встроенную функцию map
- Выведите результат в консоль.

```python
# Создаем лямбда-функцию для вычисления куба числа
cube = lambda x: x ** 3
```

```python
# Лямбда-функция для вычисления куба числа
cube = lambda x: x ** 3

#  список чисел
numbers = [1, 2, 3, 4, 5]

# Применение лямбда-функции к каждому элементу списка с использованием map
cubed_numbers = list(map(cube, numbers))

# Вывод результата 
print("Кубы чисел:", cubed_numbers)
```

Задание 10. Работа с файлами
10.1. Теория
- Расскажите, как в Python осуществляется работа с файлами
- Укажите ключевые моменты: открытие файла (open), режимы чтения/
записи, закрытие файла и контекстный менеджер (with open...)

Для открытия используется функция open(). функция принимает два основных параметра: имя файла и режим работы с ним.
например,
file = open('example.txt', 'r')  # Открытие файла для чтения

При открытии файла можно указать различные режимы:
'r' — чтение, 'w' — запись, 'a' — добавление. Данные будут добавлены в конец файла.


10.2. Практическая часть
- Напишите программу, которая создаёт (или перезаписывает, если такой
есть) файл data.txt, записывает в него несколько строк текста
(например, три строки)
- Затем откройте этот файл в режиме чтения, прочитайте содержимое и
выведите в консоль.

```python


filename = 'myLoveThings.txt'

def write():
    message = input('Введите строку: ')
    with open(filename, 'a') as file:
        file.write(message + '\n')

def read():
    with open(filename, 'r') as file:
        for message in file:
            print(message, end='') 
    print()     

 

while True:
    selection = int(input('1. Запись в файл\n2. Чтение файла\n3. Выход '))
    
    if selection == 1:
        write()
    elif selection == 2:
        read()
    
    elif selection == 3:
        break
    else:
        print('попробуйте ввести снова.')

print('Пока :3')

 
```
