```python
import networkx as nx 
import matplotlib.pyplot as plt 

G = nx.Graph() # Создаем граф

# Добавление вершин (склады и магазины)
G.add_node("Склад 1")
G.add_node("Склад 2")
G.add_node("Склад 3")
G.add_node("Магазин 1")
G.add_node("Магазин 2")

# Добавление ребер с весами (стоимость доставки)
G.add_edge("Склад 1", "Магазин 1", weight=10)
G.add_edge("Склад 1", "Склад 2", weight=15)
G.add_edge("Склад 2", "Магазин 2", weight=20)
G.add_edge("Склад 3", "Магазин 2", weight=5)
G.add_edge("Склад 3", "Склад 1", weight=25)
G.add_edge("Склад 2", "Магазин 1", weight=30)

# Поиск минимального пути от склада 1 до магазина 2
shortest_path = nx.shortest_path(G, source="Склад 1", target="Магазин 2", weight="weight")
path_len = nx.shortest_path_length(G, source="Склад 1", target="Магазин 2", weight="weight")

print(f"Минимальный путь от Cклада 1 до Магазина: {shortest_path}")
print(f"Стоимость маршрута: {path_len} единиц")

pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, "weight")

plt.figure(figsize=(8,6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Граф доставки товаров")
plt.show()
```
