1 задание.
Создайте класс Обувь. Необходимо хранить следующую информацию:
■ тип обуви;
✓мужская,
✓женская;
■ вид обуви (кроссовки, сапоги, сандалии, туфли и т.д.);
■ цвет;
■ цена;
■ производитель;
■ размер.
Создайте необходимые методы для этого класса. Реализуйте паттерн MVC для класса Обувь и код для использования модели, контроллера и представления.

```python
class Shoes:
    def __init__(self, type, view, color, price, manufacturer, size):
        self.type = type  # тип 'мужская' или 'женская'
        self.view = view  # вид обуви (кроссовки, сапоги и т.д.)
        self.color = color
        self.price = price
        self.manufacturer = manufacturer # производитель
        self.size = size

    def __str__(self): # метод для строкового представления объекта обуви
        return f" Тип: {self.type}, Вид: {self.view}, Цвет: {self.color}, Цена: {self.price}, Производитель: {self.manufacturer}, Размер: {self.size}"

    def to_dict(self): # Преобразуем объект в словарь для более удобной работы
        return {
            "type": self.type,
            "view": self.view,
            "color": self.color,
            "price": self.price,
            "manufacturer": self.manufacturer,
            "size": self.size
        }

# реализация
class ShoesModel:
    def __init__(self):
        self.shoes_list = []

    def add_shoes(self, shoes): # Метод для добавления новой обуви в список
        self.shoes_list.append(shoes)

    def get_all_shoes(self): # Метод для получения всей обуви из списка        
        return self.shoes_list

    def find_shoes_by_view(self, view):# Метод для поиска обуви по виду 
        return [shoes for shoes in self.shoes_list if shoes.view == view]

# контроллер
class ShoesController:
    def __init__(self, model):        
        self.model = model

    def add_shoes(self, type, view, color, price, manufacturer, size):
        # Метод для добавления новой обуви через модель
        new_shoes = Shoes(type, view, color, price, manufacturer, size)
        self.model.add_shoes(new_shoes)

    def get_all_shoes(self):# Метод для получения всех объектов обуви из модели        
        return self.model.get_all_shoes()

    def find_shoes_by_view(self, view):# Метод для поиска обуви по виду через модель        
        return self.model.find_shoes_by_view(view)

# Представление
class ShoesView:
    def display_shoes(self, shoes_list):# Метод для отображения списка обуви
        for shoes in shoes_list:
            print(shoes)

    def display_message(self, message):
        print(message)


   
model = ShoesModel()
controller = ShoesController(model)
view = ShoesView()

   
controller.add_shoes('мужская', 'кроссовки', 'синий', 3500, 'Nike', 43)
controller.add_shoes('женская', 'туфли', 'красный', 5000, 'Gucci', 38)

    
shoes_list = controller.get_all_shoes()
view.display_shoes(shoes_list)

    
found_shoes = controller.find_shoes_by_view('кроссовки')
view.display_message("Найденные кроссовки:")
view.display_shoes(found_shoes)

```

2 задание
Создайте класс Рецепт. Необходимо хранить следующую информацию:
■ название рецепта;
■ автор рецепта;
■ тип рецепта (первое, второе блюдо и т.д.);
■ текстовое описание рецепта;
■ ссылка на видео с рецептом;
■ список ингредиентов;
■ название кухни (итальянская, французская, украинская и т.д.).
Создайте необходимые методы для этого класса. Реализуйте паттерн MVC для класса Рецепт и код для использования модели, контроллера и представления.

```python
class Recipe:
    def __init__(self, name, author, type, description, link, ingredients, kitchen):
        self.name = name # название рецепта
        self.author = author # автор рецепта
        self.type = type # тип рецепта
        self.description = description # текстовое описание рецепта
        self.link = link # ссылка на видео
        self.ingredients = ingredients # список ингридиентов
        self.kitchen = kitchen # назвние кухни

    def __str__(self): # Строковое представление рецепта
        return f'Название рецепта: {self.name}, Автор: {self.author}, Тип рецепта: {self.type}, Кухня: {self.kitchen}'
    
    def to_dict(self): # преобразуем в словарь
        return {
        "name": self.name, 
        "author": self.author,
        "type": self.type,
        "description": self.description,
        "link": self.link,
        "ingredients": self.ingredients,
        "kitchen": self.kitchen
        }
    
# Реализация
class RecipeModel:
    def __init__(self):
        self.recipe_list = []    # список для хранения объектов

    def add_recipe(self, recipe): # метод для добавления нового рецепта в список
        self.recipe_list.append(recipe) 

    def get_all_recipes(self): # метод для получения всех рецептов из списка
        return self.recipe_list 

    def find_recipes_kitchen(self, kitchen):   # метод для поиска 
        return [recipe for recipe in self.recipe_list if recipe.kitchen == kitchen]  
    


# контроллер
class RecipeController:
    def __init__(self, model):
        self.model = model

    def add_recipe(self, name, author, type, description, link, ingredients, kitchen): # метод для добавления нового рецепта через модель
        new_recipe = Recipe(name, author, type, description, link, ingredients, kitchen)
        self.model.add_recipe(new_recipe)

    def get_all_recipes(self): # метод для получения всех рецептов из модели
        return self.model.get_all_recipes()  

    def find_recipes_kitchen(self, kitchen):  # метод для поиска рецептов из модели
        return self.model.find_recipes_kitchen(kitchen)
    

# Представление
class RecipeView:
    def display_recipes(self, recipe_list): # метод для отображения списка
        for recipe in recipe_list:
            print(recipe)

    def display_message(self, message): # метод для отображения сообщений
        print(message)    



model = RecipeModel()
controller = RecipeController(model)
view = RecipeView()

controller.add_recipe(
    name='Хинкали', 
    author='Народ из горных грузинских областей Пшави, Мтиулети и Хевсурети', 
    type='Основное блюдо',
    description='Традиционные хинкали из говядины',
    link='https://yandex.ru/video/preview/7532415065686554959',
    ingredients=['тесто', 'говядина', 'приправы'],
    kitchen='Грузинская'
)

recipe_list = controller.get_all_recipes()
view.display_recipes(recipe_list)

found_recipes = controller.find_recipes_kitchen('Грузинская')
view.display_message('Найденные грузинские рецепты:')
view.display_recipes(found_recipes)
```
