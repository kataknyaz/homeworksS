 1.Работа с GET-запросами
 Отправьте GET-запрос к API https://jsonplaceholder.typicode.com/posts и
получите список постов
 Извлеките и выведите заголовки всех постов

```python
#1 
import requests  # библиотека для работы с запросами

url = 'https://jsonplaceholder.typicode.com/posts'

# GET- получение данных с сервера
response = requests.get(url)

if response.status_code == 200:
    json_obj = response.json()
    for obj in json_obj:
        title = obj['title']
        body = obj['body']
        print(f"Название: {title}\nСодержание: {body}\n")
else:
    print(f'Ошибка: {response.status_code}')        



```

2. Получение списка задачq
 Отправьте GET-запрос к API https://jsonplaceholder.typicode.com/todos для
получения списка задач
 Выведите на экран первые 10 задач, отображая их идентификатор (id),
заголовок (title) и статус выполнения (completed)

```python
#2 
import requests  # библиотека для работы с запросами

url = 'https://jsonplaceholder.typicode.com/todos'

# GET- получение данных с сервера
response = requests.get(url)

if response.status_code == 200:
    json_obj = response.json()
    for obj in json_obj[:10]:
        id = obj['id']
        title = obj['title']
        completed = obj['completed']
        print(f"Идентификатор: {id}\nНазвание: {title}\nCтатус выполнения: {completed}\n")
else:
    print(f'Ошибка: {response.status_code}')        
```

 3. Получение списка комментариевq
 Отправьте GET-запрос к API https://jsonplaceholder.typicode.com/comments
для получения списка комментариев
 Выведите на экран первые 10 комментариев, отображая их идентификатор
(id), идентификатор поста (postId), имя (name) и тело комментария (body)

```python
#3 
import requests  # библиотека для работы с запросами

url = 'https://jsonplaceholder.typicode.com/comments'

# GET- получение данных с сервера
response = requests.get(url)

if response.status_code == 200:
    json_obj = response.json()
    for obj in json_obj[:10]:
        id = obj['id']
        postId = obj['postId']
        name = obj['name']
        body = obj['body']
        print(f"Идентификатор: {id}\nИдентификатор поста: {postId}\nИмя: {name}\nТело комментария: {body}\n")
else:
    print(f'Ошибка: {response.status_code}')        
```
