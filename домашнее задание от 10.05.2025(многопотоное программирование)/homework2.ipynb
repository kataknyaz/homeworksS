1 задание. 
При старте приложения запускаются три потока.Первый поток заполняет список случайными числами.
Два других потока ожидают заполнения. Когда список заполнен оба потока запускаются. Первый поток находит
сумму элементов списка, второй поток среднеарифметическое значение в списке. Полученный список, сумма и
среднеарифметическое выводятся на экран. 


```python
import threading
import random
import time


numbers = []  # Список для хранения случайных чисел
list_filled = threading.Event()  # Событие об уведомлении о заполнении списка

def fill_list(size):
    global numbers
    for i in range(size):        
        numbers.append(random.randint(1, 50)) # Заполняем список случайными числами от 1 до 50
        time.sleep(0.1)  # Задержка 
    list_filled.set()  # Устанавливаем событие, когда список заполнен
    print("Список заполнен:", numbers)

def calculate_sum():
    list_filled.wait()  # Ожидаем, когда список будет заполнен
    total_sum = sum(numbers)  #  сумма элементов списка
    print("Сумма элементов списка:", total_sum)

def calculate_average():
    list_filled.wait()  # Ожидаем, когда список будет заполнен
    if numbers:  # Проверяем, что список не пустой
        average = sum(numbers) / len(numbers)  # среднее значение
        print("Среднее арифметическое значение:", average)
    else:
        print("Список пуст, среднее значение не может быть рассчитано.")


size_of_list = 8  


thread_fill = threading.Thread(target=fill_list, args=(size_of_list,))
thread_sum = threading.Thread(target=calculate_sum)
thread_average = threading.Thread(target=calculate_average)

   
thread_fill.start() 
thread_sum.start()
thread_average.start()


thread_fill.join()
thread_sum.join()
thread_average.join()
```

2 задание. 
Пользователь с клавиатуры вводит путь к файлу. После чего запускаются три потока. Первый поток заполняет файл случайными числами. Два других потока ожидают заполнения. Когда файл заполнен оба потока стартуют. Первый поток находит все простые числа, второй поток факториал каждого числа в файле. Результаты поиска каждый поток должен записать в новый файл. На экран необходимо отобразить статистику выполненных
операций.


```python
import threading
import random
import time
import os


numbers = []  # Список для хранения случайных чисел
file_path = ''  # Путь к файлу
file_filled = threading.Event()  # Событие, говорящее о заполнении файла


def fill_file(path, size):     # Функция для заполнения файла случайными числами
    global numbers
    with open(path, 'w') as f:
        for i in range(size):
            num = random.randint(1, 20)  # Генерируем случайное число от 1 до 20
            f.write(f"{num}\n")  # Записываем число в файл
            numbers.append(num)  # Добавляем число в список
            time.sleep(0.1)  # Задержка 
    file_filled.set()  # Устанавливаем событие, когда файл заполнен
    print(f"Файл '{path}' заполнен.")


def find_primes(path):     # Функция для поиска простых чисел в файле
    file_filled.wait()  # Ожидаем, пока файл будет заполнен
    primes = []
    with open(path, 'r') as f:
        for line in f:
            num = int(line.strip())
            if is_prime(num):  # Проверяем, является ли число простым
                primes.append(num)
    with open('primes.txt', 'w') as f:  # Записываем простые числа в новый файл
        for prime in primes:
            f.write(f"{prime}\n")
    print(f"Простые числа записаны в 'primes.txt': {primes}")

# Функция для проверки, является ли число простым
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Функция для вычисления факториала чисел в файле
def calculate_factorials(path):
    file_filled.wait()  # Ожидаем, пока файл будет заполнен
    factorials = []
    with open(path, 'r') as f:
        for line in f:
            num = int(line.strip())
            fact = factorial(num)  # Вычисляем факториал
            factorials.append(fact)
    with open('factorials.txt', 'w') as f:  # Записываем факториалы в новый файл
        for fact in factorials:
            f.write(f"{fact}\n")
    print(f"Факториалы записаны в 'factorials.txt': {factorials}")

# Функция для вычисления факториала числа
def factorial(n):
    if n == 0 or n == 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result


size_of_list = int(input("Введите количество случайных чисел для записи в файл: "))
file_path = input("Введите путь к файлу (например, 'numbers.txt'): ")


thread_fill = threading.Thread(target=fill_file, args=(file_path, size_of_list))
thread_primes = threading.Thread(target=find_primes, args=(file_path,))
thread_factorials = threading.Thread(target=calculate_factorials, args=(file_path,))
thread_fill.start()  # Запускаем поток для заполнения файла
thread_primes.start()  # Запускаем поток для поиска простых чисел
thread_factorials.start()  # Запускаем поток для вычисления факториалов


thread_fill.join()
thread_primes.join()
thread_factorials.join()
print("Все операции завершены.")
```

3 задание. 
Пользователь с клавиатуры вводит путь к существующей директории и к новой директории. После чего
запускается поток, который должен скопировать содержимое директории в новое место. Необходимо сохранить
структуру директории. На экран необходимо отобразить статистику выполненных операций.

```python
import os
import shutil
import threading

# Копирует содержимое директории src в директорию dst.
def copy_directory(src, dst):
  
    total_files = 0  # Общее количество файлов
    total_dirs = 0   # Общее количество директорий

    for dirpath, dirnames, filenames in os.walk(src):  
        
        relative_path = os.path.relpath(dirpath, src)  
        new_dir = os.path.join(dst, relative_path)  # Создаем новый путь
        os.makedirs(new_dir, exist_ok=True)  # Создаем директорию, если она не существует
        total_dirs += 1  

        for filename in filenames:  # Проходим по всем файлам в текущей директории
            src_file = os.path.join(dirpath, filename)  # путь к исходному файлу
            dst_file = os.path.join(new_dir, filename)  #  путь к новому файлу
            shutil.copy2(src_file, dst_file)  
            total_files += 1  # Увеличиваем счетчик файлов

    print(f"Копирование завершено. Скопировано {total_files} файлов и {total_dirs} директорий.")

def start_copy_thread(src, dst):
    """Запускает поток для копирования директории."""
    copy_thread = threading.Thread(target=copy_directory, args=(src, dst))
    copy_thread.start()  # Запускаем поток

if __name__ == "__main__":
    source_directory = input("Введите путь к существующей директории: ")
    destination_directory = input("Введите путь к новой директории: ")

    if os.path.exists(source_directory) and os.path.isdir(source_directory):
        start_copy_thread(source_directory, destination_directory)
    else:
        print("Указанный путь не существует или не является директорией.")
```

4 задание. 
Пользователь склавиатуры вводит путь к существующей директории и слово для поиска. После чего запускаются
два потока. Первый должен найти файлы, содержащие искомое слово и слить их содержимое в один файл. Второй поток ожидает завершения работы первого потока. После чего проводит вырезание всех запрещенных слов (список этих слов нужно считать из файла с запрещенными словами) из полученного файла. На экран необходимо отобразить статистику выполненных операций.

```python
import os
import threading
from queue import Queue

# Функция для поиска файлов и объединения их содержимого
def find_and_merge_files(directory, search_word, output_file):
    with open(output_file, 'w', encoding='utf-8') as outfile:
        for root, _, files in os.walk(directory):
            for file in files:
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r', encoding='utf-8') as infile:
                        if search_word in infile.read():
                            infile.seek(0)  # Вернуться в начало файла
                            outfile.write(infile.read() + "\n")
                            print(f"Файл {file_path} добавлен в {output_file}")
                except Exception as e:
                    print(f"Ошибка при чтении файла {file_path}: {e}")

# Функция для удаления запрещенных слов
def remove_forbidden_words(input_file, forbidden_words_file, output_file):
    with open(forbidden_words_file, 'r', encoding='utf-8') as f:
        forbidden_words = set(word.strip() for word in f.readlines())

    with open(input_file, 'r', encoding='utf-8') as infile:
        content = infile.read()

    for word in forbidden_words:
        content = content.replace(word, '')

    with open(output_file, 'w', encoding='utf-8') as outfile:
        outfile.write(content)
    
    print(f"Запрещенные слова удалены. Результат сохранен в {output_file}")

# Главная функция
def main():
    directory = input("Введите путь к директории: ")
    search_word = input("Введите слово для поиска: ")
    forbidden_words_file = input("Введите путь к файлу с запрещенными словами: ")
    merged_file = "merged.txt"
    cleaned_file = "cleaned.txt"

    # Создание потоков
    find_thread = threading.Thread(target=find_and_merge_files, args=(directory, search_word, merged_file))
    find_thread.start()
    
    # Ожидание завершения первого потока
    find_thread.join()

    # Запуск второго потока для удаления запрещенных слов
    remove_thread = threading.Thread(target=remove_forbidden_words, args=(merged_file, forbidden_words_file, cleaned_file))
    remove_thread.start()
    remove_thread.join()

    print("Все операции завершены.")

if __name__ == "__main__":
    main()
```
